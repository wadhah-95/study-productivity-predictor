```python
import pandas as pd 
import matplotlib as plt
df=pd.read_csv("../data/study-sessions.csv")
df.head()
df.shape
df.columns
df.info()
df.describe()
df.groupby("productive")[["sleep_hours", "phone_usage_hours", "energy_level", "mood_level"]].mean()
df.isnull().sum()
df["productive"].value_counts().plot(kind="bar", title="productive vs Non-productive sessions")
plt.xlabel("Productive")
plt.ylabel("Count")
plt.show()
```
